```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
import numpy as np
%matplotlib inline
#Пункт А.
#Задаем временной интервал:
t = np.arange(0, 6, 0.01)
#Вводим амплитуды колебаний:
A_1 = 0.1
A_2 = 0.5
#Фазы колебаний:
fi_1 = 0
fi_2 = math.pi/2
#Частота:
w = 3
#Уравнения движения маятника для каждого колебания:
x1 = A_1 * np.sin(w * t + fi_1)
x2 = A_2 * np.sin(w * t + fi_2)
#Суммарное колебание:
x = x1 + x2
#Строим график зависимости смещения маятника х от времени t:
fig = plt.figure()
ax = plt.axes()
ax.plot(t, x1, '-g', label='x1', linewidth = 3, color='yellow')
ax.plot(t, x2, '-g', label = 'x2', linewidth = 3, color = 'orange')
ax.plot(t, x, '--g', label = 'x1+x2', linewidth = 3, color = 'red')
ax.legend(loc='lower left', fancybox = True, framealpha=1, frameon=True)
ax.grid(color='lightgreen', linewidth=1,linestyle='--')
plt.xlabel('время t, c')
plt.ylabel('смещение маятника х, м')
```

```python
#Пункт B.
#Задаем частоты складываемых колебаний:
w_1 = 45
w_2 = 48
#Задаем амплитуду:
A = 1
#Задаем уравнения движения для каждого колебания:
x_1 = A * np.sin(w_1 * t)
x_2 = A * np.sin(w_2 * t)
#Сумма колебаний:
x_ = x_1 + x_2
#Уравнение огибающей:
x_0 = 2 * A *np.cos((w_2 - w_1)*t/2)
#Строим график зависимости смещения маятника х от времени t:
fig = plt.figure()
ax = plt.axes()
ax.plot(t, x_, '-g', label='x_1+x_2', linewidth = 3, color='green')
ax.plot(t, x_0, '--g', label = 'огибающая', linewidth = 3, color = 'blue')
ax.legend(loc='lower left', fancybox = True, framealpha=1, frameon=True)
ax.grid(color='lightgreen', linewidth=1,linestyle='--')
plt.xlabel('время t, c')
plt.ylabel('смещение маятника х, м')
```

```python
#Пункт C.
#Задаем временной диапазон:
t1 = np.arange(0,7, 0.01)
#Задаем частоты складываемых колебаний:
w1 = 1
w2 = 2
#Задаем амплитуду:
A_ = 0.2
#Задаем уравнения движения для каждого колебания:
x__1 = A_ * np.sin(w1 * t1 + fi_1)
x__2 = A_ * np.sin(w2 * t1 + fi_2)
#Строим фигуру Лиссажу:
fig = plt.figure()
ax = plt.axes()
ax.plot(x__1, x__2, linewidth = 3, color='red')
ax.grid(color='lightgreen', linewidth=1,linestyle='--')
plt.xlabel('смещение маятника х, м')
plt.ylabel('смещение маятника х, м')
```
