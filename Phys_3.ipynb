```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
import numpy as np
from mpl_toolkits import mplot3d
%matplotlib inline

#Пункт А.
#Задаем временной диапазон:
t = np.linspace(0,6)
#Начальная амплитуда:
A = 0.08
#Коэффициент жесткости пружины:
k = 0.5
#Начальное значение массы маятника:
m0=0.6
#Находим начальную частоту колебаний:
w0 = math.sqrt(k/m0)
#Задаем уравнение движения точки (cos - т.к. начало движение из максимального отклонения от положения равновесия):
x = A * np.cos(w0*t)
#Строим график зависимости координаты от времени:
fig = plt.figure()
ax = plt.axes()
ax.plot(t,x, linewidth = 3, color='green')
ax.grid(color='lightgreen', linewidth=1,linestyle='--')
plt.xlabel('время t, c')
plt.ylabel('координата точки х, м')
```

```python
#Вычисляем скорость движения:
v = A * w0 * np.sin(w0 * t)
#Строим графики зависимости скорости от времени:
fig = plt.figure()
ax = plt.axes()
ax.plot(t, v, linewidth = 3, color='blue')
ax.grid(color='lightgreen', linewidth=1,linestyle='--')
plt.xlabel('время t, c')
plt.ylabel('скорость точки v, м/c')
```

```python
#Находим ускорение:
a = -A * w0 * w0 * np.cos(w0 * t)
#Строим графики зависимости ускорения от времени:
fig = plt.figure()
ax = plt.axes()
ax.plot(t, a, linewidth = 3, color='red')
ax.grid(color='lightgreen', linewidth=1, linestyle='--')
plt.xlabel('время t, c')
plt.ylabel('ускорение a, м/c2')
```

```python
#Пункт B.
#Задаем диапазон изменения массы (от m0 до 3*m0):
m = np.linspace(m0, 3*m0, 3)
#Рассчитываем частоты для масс 0.6, 1.2, 1.8 кг:
w = np.sqrt(k/m)
w1 = np.sqrt(k/m0)
w2 = np.sqrt(k/(2*m0))
w3 = np.sqrt(k/(3*m0))
#Находим уравнения движения для масс 0.6, 1.2, 1.8 кг:
x1 = np.cos(w1 * t)
x2 = np.cos(w2 * t)
x3 = np.cos(w3 * t)
#Строим график зависимости координаты х от времени t для m=0.6, m=1.2, m=1.8 кг:
fig = plt.figure()
ax = plt.axes()
ax.plot(t, x1, '-g', label='m = 0.6', linewidth = 3, color='blue')
ax.plot(t, x2, '-g', label = 'm = 1.2', linewidth = 3, color = 'orange')
ax.plot(t, x3, '-g', label = 'm = 1.8', linewidth = 3, color = 'green')
ax.legend(loc='lower left', fancybox = True, framealpha=1, frameon=True)
ax.grid(color='lightgreen', linewidth=1,linestyle='--')
plt.xlabel('время t, c')
plt.ylabel('координата точки х, м')
```

```python
#Строим трехмерный график:
def f(m, t):
    return np.cos(w*t)
X,Y = np.meshgrid(m, t)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=2, cmap='RdBu', edgecolor='none')
ax.set_xlabel('m')   
ax.set_ylabel('t')
ax.set_zlabel('x')
ax.locator_params(tight=None, nbins=7)
ax.view_init(45,165)
```

```python
#Пункт С.
#Задаем значения вязкого трения:
r = np.array([0.07, 0.09, 0.1])
r1 = 0.07
r2 = 0.09
r3 = 0.1
#Рассчитываем дельта:
delta = r/(2*m0)
delta1 = r1/(2*m0)
delta2 = r2/(2*m0)
delta3 = r3/(2*m0)
#Задаем уравнения движения для r1, r2, r3:
x_1 = A * np.cos(t * w0) * np.exp(-t*delta1)
x_2 = A * np.cos(t * w0) * np.exp(-t*delta2)
x_3 = A * np.cos(t * w0) * np.exp(-t*delta3)
#Строим график зависимости координаты х от времени t для k=0.07, k=0.09, k=0.1 (рисунок 3-6):
fig = plt.figure()
ax = plt.axes()
ax.plot(t, x_1, '-g', label='k = 0.07', linewidth = 3, color='green')
ax.plot(t, x_2, '-g', label = 'k = 0.09', linewidth = 3, color = 'orange')
ax.plot(t, x_3, '-g', label = 'k =0.1', linewidth = 3, color = 'red')
ax.legend(loc='lower left', fancybox = True, framealpha=1, frameon=True)
ax.grid(color='lightgreen', linewidth=1,linestyle='--')
plt.xlabel('время t, c')
plt.ylabel('координата точки х, м')
```

```python
#Строим трехмерный график для k=0.07, k=0.09, k=0.1:
def f(r, t):
    return A * np.cos(t*w0)*np.exp(-t*delta)
X1, Y1 = np.meshgrid(r, t)
Z1 = f(X1, Y1)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X1, Y1, Z1, rstride=1, cstride=2, cmap='jet', edgecolor='none')
ax.set_xlabel('r')   
ax.set_ylabel('t')
ax.set_zlabel('x')
ax.locator_params(tight=None, nbins=5)
ax.view_init(35,145)
```

```python
#Пункт D.
#Задаем дельта:
delta0 = delta1
#Задаем частоты вынуждающей силы:
w_1 = 1
w_2 = 2
w_3 = 3
#Рассчитываем амплитуды:
A_1 = 1/math.sqrt((w0*w0 - w_1*w_1)*(w0*w0 - w_1*w_1 + 4*delta0*delta0*w_1*w_1))
A_2 = 1/math.sqrt((w0*w0 - w_2*w_2)*(w0*w0 - w_2*w_2 + 4*delta0*delta0*w_2*w_2))
A_3 = 1/math.sqrt((w0*w0 - w_3*w_3)*(w0*w0 - w_3*w_3 + 4*delta0*delta0*w_3*w_3))
#Рассчитываем значения вынуждающей силы:
F1 = math.atan(2*delta0*w_1/(w0*w0 - w_1*w_1))
F2 = math.atan(2*delta0*w_2/(w0*w0 - w_2*w_2))
F3 = math.atan(2*delta0*w_3/(w0*w0 - w_3*w_3))
#Находим уравнение движения:
x__1 = A_1*np.cos(t*w_1 + F1)
x__2 = A_2*np.cos(t*w_2 + F2)
x__3 = A_3*np.cos(t*w_3 + F3)
#Строим график зависимости координаты х от времени t при вынужденных колебаниях (рисунок 3-8):
fig = plt.figure()
ax = plt.axes()
ax.plot(t, x__1, '-g', label='частота = 1(1/c)', linewidth = 3, color='yellow')
ax.plot(t, x__2, '-g', label = 'частота = 2(1/c)', linewidth = 3, color = 'orange')
ax.plot(t, x__3, '-g', label = 'частота = 3(1/c)', linewidth = 3, color = 'red')
ax.legend(loc='lower left', fancybox = True, framealpha=1, frameon=True)
ax.grid(color='lightgreen', linewidth=1,linestyle='--')
plt.xlabel('время t, c')
plt.ylabel('координата точки х, м')
```

```python
#Строим трехмерный график для w=1, w=2, w=3:
u = np.linspace(1,3,3)
t = np.linspace(0,6)
A_ = 1/np.sqrt((w0*w0-u*u)*(w0*w0-u*u+4*delta0*delta0*u*u))
F = np.arctan(2*delta0*u/(w0*w0-u*u))
def f(u,t):
    return A_ * np.cos(t*u + F)
X2, Y2 = np.meshgrid(u,t)
Z2 = f(X2, Y2)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X2, Y2, Z2, rstride=1, cstride=2, cmap='viridis', edgecolor='none')
ax.set_xlabel('частота')   
ax.set_ylabel('время')
ax.set_zlabel('координата')
ax.locator_params(tight=None, nbins=5)
ax.view_init(45,125)
```
