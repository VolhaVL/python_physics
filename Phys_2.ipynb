```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
import numpy as np
plt.style.use('seaborn-whitegrid')
%matplotlib inline
#Задаем временной диапазон:
t = np.linspace(0,0.5)
#Вводим уравнение для ускорения:
a = (4 - 9*(t**2))**(-0.5)
#Строим график зависимости ускорения от времени:
fig = plt.figure()
ax = plt.axes()
ax.plot(t, a, linewidth = 3, color='green')
ax.grid(color='lightgreen', linewidth=1,linestyle='--')
plt.xlabel('время t, c')
plt.ylabel('ускорение точки а, м/с2')
```

```python
#Вычисляем интеграл:
from scipy.integrate import odeint
def first(v, t):   
    dvdt = (4-9*(t**2))**(-0.5)
    return dvdt
v = odeint(first, 0, t)
#Строим график зависимости скорости от времени для теоретического и рассчитанного значений:
fig = plt.figure()
ax = plt.axes()
ax.plot(t, np.arcsinh(1.5*t)/3, '-g', label='теор', linewidth = 3, color='blue')
ax.plot(t, v, '-g', label = 'практ', linewidth = 3, color = 'orange')
ax.legend(loc='upper left', fancybox = True, framealpha=1, frameon=True)
ax.grid(color='lightgreen', linewidth=1,linestyle='--')
```
