```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
import numpy as np
#Задаем стиль отображения графиков:
plt.style.use('seaborn-whitegrid')
%matplotlib inline
#Определяем временной диапазон:
dt = 0.02
t = np.linspace(0,1)
#Вводим уравнение для координаты:
x = 2 * math.e**(-3*t)
#Находим скорость(ускорение) как производную от х(v):
v = np.diff(x)/dt
a = np.diff(v)/dt
vt = (-6) * math.e**(-3*t)
at = 18 * math.e**(-3*t)
#Строим график зависимости координаты от времени:
fig = plt.figure()
ax = plt.axes()
ax.plot(t,x, linewidth = 3, color='green')
ax.grid(color='lightgreen', linewidth=1,linestyle='--')
plt.xlabel('время t, c')
plt.ylabel('координата точки х, м')
   
```
